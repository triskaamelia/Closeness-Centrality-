```python
#library yang dibutuhkan
%matplotlib inline

import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
```

```python
#membaca data
df2 = pd.read_csv('Salinan dari TAGS v6.1.9.1 - Archive.csv')
```

```python
#mengambil data id user dari dan menuju untuk plotting
data2 = df2[["id_str", "from_user_id_str"]]
df = pd.DataFrame(data2)
```

```python
#plotting graph node dan edge (membuat grafik)
twitter_graph = nx.from_pandas_edgelist(df, source="id_str", target="from_user_id_str")
nx.info(twitter_graph)
```

```python
#menampilkan hasil graph node dan edge
plt.figure(figsize=(65,65))
nx.draw_networkx(twitter_graph)
plt.show()
plt.savefig("twitter_graph.png")
```

```python
#mengecek banyak edge per node
nx.degree(twitter_graph)
```

```python
#mencari node dgn nilai closeness centrality terbesar
close_dict = nx.closeness_centrality(twitter_graph)
closeness = pd.DataFrame(list(close_dict.items()),columns=['node','closeness'])

x = closeness.iloc[0,1]
for i in closeness:
    j=1
    y = closeness.iloc[j,1]
    if x<y:
        x = y
    id_x = closeness.iloc[j,0]
    j+=1
display(id_x,x)
```
